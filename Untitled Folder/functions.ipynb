```python
#calculator
def calc(x,y,ch):
    if ch == '+':
        r = x+y
    elif ch == '-':
        r = x-y
    elif ch == '*':
         r = x-y
    else:
        if y != 0:
            if ch == '/':
                r = x/y
            elif ch == '%':
                r = x%y
            else:
                r = print('WRONG INPUT')
        else:
            r = print('Y should not be zero')
    return r
```

```python
#area_calculation
from math import sqrt, pi
def area(r,b=None,c=None):
    if b and c:
        s = (r + b + c)/2
        result = sqrt(s*(s-r)*(s-b)*(s-c))
    elif b:
        result = r*b
    else:
        result = r*r*pi
    return result
    
```

```python
area(1,2,2)
```

```python
#multiple_arguments
def square(*args):
    print(args)
    r = []
    for var in args:
        r.append(var**2)
    return r
```
